```python
from utils import load_json

data_train = load_json("data/train_nlu.json")
data_test = load_json("data/test_nlu.json")
```

```python
## Plot Train data
from collections import Counter
import matplotlib.pyplot as plt

intents_train = [ sample['intent_name'] for sample in data_train]

intents = Counter(intents_train)
plt.figure(figsize=(16,10))
plt.bar(intents.keys(),intents.values())
plt.xticks(rotation = 90)
plt.show()

```

```python
## Plot Train data
from collections import Counter
import matplotlib.pyplot as plt

intents_test = [ sample['intent_name'] for sample in data_test]

intents = Counter(intents_test)
plt.figure(figsize=(16,10))
plt.bar(intents.keys(),intents.values())
plt.xticks(rotation = 90)
plt.show()
```
